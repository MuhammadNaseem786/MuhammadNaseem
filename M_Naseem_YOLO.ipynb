```python
# Load the Yolo Model

from ultralytics import YOLO

# Load the model (pretrained weights)
model = YOLO('yolov8n.pt')  # YOLOv8n is a lightweight version
```

```python
# Train on COCO128 dataset
model.train(data='coco128.yaml', epochs=10, imgsz=640)  # Adjust epochs and image size as needed
```

```python
# Evaluate the model on validation data
metrics = model.val()
```

```python
# Test the trained model on an example image
!wget https://ultralytics.com/images/zidane.jpg -O test.jpg  # Download test image
results = model.predict(source="test.jpg", save=True, conf=0.5)  # Run inference
```
