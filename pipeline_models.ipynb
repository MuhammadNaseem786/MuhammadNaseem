```python
from transformers import pipeline
```

```python
clasifier = pipeline("sentiment-analysis")
```

```python
clasifier("we are very happy")
```

```python
text1 = "how are you i am good"
```

```python
clasifier(text1)
```

```python
text1 = "before the partision kolachi is the name of karachi"
```
